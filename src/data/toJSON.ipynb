```python
import pandas as pd
import json
```

```python
excel_data_df = pd.read_excel('words.xlsx')
json_str = excel_data_df.to_json(orient='columns')
json_dict = json.loads(json_str)
json_dict = json_dict["Unnamed: 0"]
print(json_dict)
```

```python
# filter unwanted keys
final_dict = dict(filter(lambda item: item[1] is not None, json_dict.items()))
final_dict = dict(filter(lambda item: len(item[1]) <= 5, final_dict.items()))

print(final_dict)
```

```python
with open('data.json', 'w') as json_file:
    json.dump(final_dict, json_file)
```

```python
# convert dictionary into array of words
lst = list(final_dict.values())
print(lst)
```

```python
# get scores for every letter
from collections import Counter
letter_lst = [i for j in lst for i in j]
print(letter_lst)

d = dict(Counter(letter_lst))
print(d)
```

```python
# determine max score
letter_scores = list(d.values())
max_score = max(letter_scores) * 5
print(letter_scores)
print(max_score)

# get scores for every word
def frequency_score(word, scores_dict, max_score):
    scores_array = []
    for i in word:
        scores_array.append(scores_dict.get(i))

    scores_array = sum(scores_array) / max_score
    return scores_array

print(frequency_score('aback', d, max_score))
```

```python
# get dissimilary score
def dissim_score(word):
    word_set = set(word)
    word_score = len(word_set) / len(word)
    return word_score

print(dissim_score('aback'))
```
