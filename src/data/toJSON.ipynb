```python
import pandas as pd
import json
```

```python
excel_data_df = pd.read_excel('words.xlsx')
json_str = excel_data_df.to_json(orient='columns')
json_dict = json.loads(json_str)
json_dict = json_dict["Unnamed: 0"]
print(json_dict)
```

```python
# filter unwanted keys
final_dict = dict(filter(lambda item: item[1] is not None, json_dict.items()))
final_dict = dict(filter(lambda item: len(item[1]) <= 5, final_dict.items()))

print(final_dict)
```

```python
with open('data.json', 'w') as json_file:
    json.dump(final_dict, json_file)
```
